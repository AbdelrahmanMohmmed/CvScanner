```python
import pandas as pd
import pickle
```

```python
people = pd.read_csv("D:/FCDS/A.courses/Training Arabian/archive/1_people.csv")
people.head()
```

```python
abilities = pd.read_csv("D:/FCDS/A.courses/Training Arabian/archive/2_abilities.csv")
abilities.head()
```

```python
education = pd.read_csv("D:/FCDS/A.courses/Training Arabian/archive/3_education.csv")
education.head()
```

```python
experience = pd.read_csv("D:/FCDS/A.courses/Training Arabian/archive/4_experience.csv")
experience.head()
```

```python
person_skills = pd.read_csv("D:/FCDS/A.courses/Training Arabian/archive/5_person_skills.csv")
person_skills.head()
```

```python
skills = pd.read_csv("D:/FCDS/A.courses/Training Arabian/archive/6_skills.csv")
skills.head(10)
```

```python
agg_abilities = abilities.groupby('person_id')['ability'].agg(lambda x: ','.join(x.dropna().unique())).reset_index(name='abilities')
agg_education = education.groupby('person_id')['program'].agg(lambda x: ','.join(x.dropna().unique())).reset_index(name='programs')
agg_experience = experience.groupby('person_id')['title'].agg(lambda x: ','.join(x.dropna().unique())).reset_index(name='job_titles')
agg_skills = person_skills.groupby('person_id')['skill'].agg(lambda x: ','.join(x.dropna().unique())).reset_index(name='skills')

result = (people[['person_id', 'name']]
          .merge(agg_abilities, on='person_id', how='left')
          .merge(agg_education, on='person_id', how='left')
          .merge(agg_experience, on='person_id', how='left')
          .merge(agg_skills, on='person_id', how='left')
          )
```

```python
result.to_csv('clean_data.csv', index=False)
```
